# Lightweight Corpus Size Analysis 

```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 

sns.set_context('talk')
sns.set_style('whitegrid')
```

```python
# Load the sizes data from the CSV
data = pd.read_csv('../fixtures/sizes.csv', names=['category', 'fileid', 'nbytes'])
data.describe()
```

```python
g = sns.violinplot('nbytes', y='category', data=data)
g.set_title('Violin Plots of Document Sizes (bytes) by Category')
```

```python
g = sns.violinplot('nbytes', y='category', data=data[data.category!='essays'])
g.set_title('Violin Plots of Document Sizes (bytes) by Category Excluding Essays')
```

```python
g = sns.barplot('nbytes', y='category', data=data)
g.set_title("Mean filesize in bytes by category")
```

```python
g = sns.barplot('nbytes', y='category', data=data[data.category!='essays'])
g.set_title("Mean filesize in bytes by category excluding essays")
```

```python
g= sns.boxplot('nbytes', y='category', data=data)
g.set_title('Boxplots of Document Sizes (in bytes) by Category')
```

```python
g= sns.boxplot('nbytes', y='category', data=data[data.category!='essays'])
g.set_title('Boxplots of Document Sizes (in bytes) by Category Excluding Essays')
```

```python
g= sns.boxplot('nbytes', y='category', data=data[data.nbytes<0.2e7])
g.set_title('Boxplots of Document Sizes (in bytes) by Category Excluding Large Sizes')
```
